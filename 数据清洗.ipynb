```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
dt = pd.read_excel('1_115_200000_300000_1.xlsx')
dt1 = dt.drop(index=0)
dt2 = dt1.drop(columns=['test_results', 'note'])
data = dt2.dropna()
data
```

```python
from pymongo import MongoClient
```

```python
client = MongoClient('localhost:27017')
mongo = client['zhejiang']
coll = mongo['200k_300k']
```

```python
for i in data.index:
    df = data.loc[i]
    item = {}
    item['corpName'] = df['manufacturing_name']
    item['address'] = df['manufacturing_addr']
    item['corpNameBy'] = df['company_name']
    item['addressBy'] = df['province_name']
    item['addressByRegionId'] = int(df['province'])
    item['commodityName'] = df['food_name']
    item['createDate'] = df['batch']
    item['fl'] = df['fl']
    item['ggh'] = df['notice']
    item['ggrq'] = df['notice_data']
    item['rwly'] = df['source']
    item['model'] = df['model']
    item['next'] = ''
    coll.insert_one(item)
```
